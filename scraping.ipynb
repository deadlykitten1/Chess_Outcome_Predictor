Now we have a list of table entries in html that contains data about each of Magnus's games (on this page at least)
So, lets loop through this list and extract the important data into a dataframe 

Refactoring to make looping through the different urls easier

```python
html_data = []

for num in range(0, 60):
    URL = f"https://www.chess.com/games/search?opening=&openingId=&p1=Magnus+Carlsen&p2=&mr=&lsty=1&year=&lstMoves=1&moves=&fen=&sort=page={num}"
    html_data.extend(cook_chess_soup(URL))
```

```python
white_player_list = []
white_player_rating_list = []
black_player_list = []
black_player_rating_list = []
result_list = []
game_length_list = []
year_list = []
game_list = []

for chunk in html_data:
    white_player_list.append(str(chunk.find_all('td')[0].find_all('a')[0].find_all('div')[0].find_all('span')[0].string))
    white_player_rating_list.append(str(chunk.find_all('td')[0].find_all('a')[0].find_all('div')[0].find_all('span')[1].string))
    black_player_list.append(str(chunk.find_all('td')[0].find_all('a')[0].find_all('div')[1].find_all('span')[0].string))
    black_player_rating_list.append(str(chunk.find_all('td')[0].find_all('a')[0].find_all('div')[1].find_all('span')[1].string))
    result_list.append(str(chunk.find_all('td')[1].find_all('a')[0].string))
    game_length_list.append(str(chunk.find_all('td')[2].find_all('a')[0].string))
    year_list.append(chunk.find_all('td')[3].a['title'])
    game_list.append(str(chunk.find_all('td')[0].find_all('a')[1].find_all('span')[1].text))

```

Now that we have all the lists of data, build the dataframe

```python
games_df = pd.DataFrame()


#There is probably a better way to do this but whatever
games_df['white_player'] = white_player_list
games_df['white_player_rating'] = white_player_rating_list
games_df['black_player'] = black_player_list
games_df['black_player_rating'] = black_player_rating_list
games_df['result'] = result_list
games_df['length'] = game_length_list
games_df['year'] = year_list
games_df['game'] = game_list
```
